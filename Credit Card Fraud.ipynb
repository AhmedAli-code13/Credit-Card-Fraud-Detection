```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('creditcard.csv')
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace= True, ignore_index= True)
```

```python
df.isna().sum()
```

# Data Preprocessing

### Split Data into Input Features and Target Column

```python
x = df.drop('Class', axis= 1)
y = df['Class']
```

```python
y.value_counts()
```

```python
fig = px.histogram(df, x="Class", color= 'Class', title='Amount distribution by Class')
fig.show()
```

### Split Data into Train & Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42, stratify= y)
```

```python
scaling_cols = x_train.columns
scaling_cols
```

```python
from sklearn.preprocessing import RobustScaler

rc = RobustScaler()

x_train[scaling_cols] = rc.fit_transform(x_train)
x_test[scaling_cols] = rc.transform(x_test)
```

```python
x_train.head()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state= 42)

x_train, y_train = smote.fit_resample(x_train, y_train)
```

```python
y_train.value_counts()
```

### Machine Learning

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

lr = LogisticRegression(random_state= 42, n_jobs= -1)

lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)
y_prob = lr.predict_proba(x_test)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state= 42, n_jobs= -1)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
y_prob = rf.predict_proba(x_test)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
```

#### XGBoost

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(random_state = 42 )

xgb.fit(x_train, y_train)

y_pred = xgb.predict(x_test)
y_prob = xgb.predict_proba(x_test)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
```
